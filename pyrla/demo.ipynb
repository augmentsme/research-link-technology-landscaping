```python
import pandas as pd
```

```python
grants = pd.read_json("grants.json")
```

```python
nhmrc_grants = grants[grants['source'] == 'nhmrc.org']
```

```python
nhmrc_grants.columns
```

```python
nhmrc_grants.start_date.map(pd.Timestamp).dt.year.value_counts().sort_index().plot(kind='bar', title='Number of NHMRC grants by year')
```
